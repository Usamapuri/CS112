```r
lm1 <- lm(foo$nowtot ~ foo$Dems + foo$Repubs + foo$Christian +
         foo$age + foo$srvlng + foo$demvote + foo$hasgirls, data = foo)

summary(lm1)
#The treatment effect is -0.4523
#Standard error of 1.903638
```

```r
#Checking the balance
X <- lm1$fitted.values
Y <- foo$nowtot
Tr <- foo$hasgirls

mtch <- Match(Y=Y, Tr=Tr, X=X, M=1)
MatchBalance(hasgirls ~ Dems +Repubs + Christian + age  + srvlng + demvote, data=foo, match.out = mtch)
# Minimum pvalue: 0.0020402 for $age
```

```r
##GENETIC MATCHING
library(rgenoud)
set.seed(2324)

X = cbind(foo$Dems,foo$Repubs , foo$Christian , foo$age  , foo$srvlng , foo$demvote)


genout <- GenMatch(Tr = foo$hasgirls, X = X, pop.size = 20, nboots=250, 
                   estimand="ATT",unif.seed = 123, int.seed = 92485)


mout1 <- Match(Tr= foo$hasgirls, X= X, Weight.matrix = genout)
MatchBalance(foo$hasgirls ~ foo$Dems +foo$Repubs + foo$Christian + foo$age  + foo$srvlng + foo$demvote,
             match.out = mout1, nboots = 250)
```

```r
#MatchBalance with Y
Y= foo$nowtot
mout2 <- Match(Tr= foo$hasgirls, X= X, Y=Y, Weight.matrix = genout)

MatchBalance(foo$hasgirls ~ foo$Dems +foo$Repubs + foo$Christian + foo$age  + foo$srvlng + foo$demvote,
             match.out = mout2, nboots = 250)

print ("The estimated treatment effect:")
mout2$est [,1]
print ("The standard error:")
mout2$se
print ("The confidence interval for the treatment effect:")
c(mout2$est + mout2$se*-1.96, mout2$est + mout2$se*1.96)
```

```r
##############################
#####     QUESTION 3      ####
##############################
library(agridat)
library(rbounds)
library(Matching)
library(sensitivitymv)

data(diggle.cow)

str(diggle.cow)

# Remove NA value
foo <- diggle.cow[-which(is.na(diggle.cow$weight)==TRUE),]

# Turn iron and infection treatment into true and false
foo$iron <- foo$iron == "Iron"
foo$infect <- foo$infect == "Infected"

lm.iron <- lm(weight ~ iron, data=foo)
lm.infe <- lm(weight ~ infect, data=foo)
summary(lm.iron)
summary(lm.infe)

ps.glm.iron <- glm(iron ~ day, data=foo, family=binomial)
ps.glm.infe <- glm(infect ~ day, data=foo, family=binomial)

Tr.iron <- foo$iron
Tr.infe <- foo$infect

Y <- foo$weight

mout.iron  <- Match(Y=Y, Tr=Tr.iron, X=ps.glm.iron$fitted, ties=FALSE)
mout.infe  <- Match(Y=Y, Tr=Tr.infe, X=ps.glm.infe$fitted, ties=FALSE)

### Using Psens
psens(mout.iron, Gamma = 3, GammaInc = 0.1)
psens(mout.infe, Gamma = 9, GammaInc = 0.1)

### USING SENMV MATCH
make_ymat<-function(matches, y){
  
  # Get indices of treated and control units
  treated <- unique(matches[, 1]) # Remove repeated indices
  controls <- matches[, 2]        # Keep repeated indices
  
  # Get outcomes of control units
  y_ctrls <- y[controls]
  
  # Create a table to check how many matches for each control  
  trt_table <- table(treated)
  
  # Get number of sets and size of largest set
  n_sets <- length(trt_table)
  max_ctrls <- max(trt_table)
  
  # Smallest table necessary is number of sets vs. size of largest set
  y_mat <- matrix(NA, n_sets, max_ctrls + 1)
  
  m <- 0 # Auxiliary indexer that will run linearly through the matches sets
  
  # For each set
  for (i in 1:n_sets){
    
    y_mat[i, 1] <- y[treated[i]] # Get treated outcome
    
    # Get controls outcomes
    y_mat[i, 2:(1+trt_table[i])] <- y_ctrls[(m+1):(m+trt_table[i])]
    
    # Advance the indexer
    m <- m + trt_table[i]
  }
  y_mat
}

# IRON TREATMENT
genout.iron <- GenMatch(Tr=Tr.iron, X=ps.glm.iron$fitted, estimand="ATT", M=1,
                   pop.size=100, max.generations=10, wait.generations=1, print.level=0, ties=FALSE)
ymat <- make_ymat(genout.iron$matches, Y) 
for (gamma in seq(1, 2, 0.1)) 
  cat(sprintf("Gamma: %.1f\tp-val: %.5f\n", 
              gamma, senmv(y=ymat, gamma=gamma, method='t')$pval))

# Infection TREATMENT
genout.infe <- GenMatch(Tr=Tr.infe, X=ps.glm.infe$fitted, estimand="ATT", M=1,
                        pop.size=100, max.generations=10, wait.generations=1, print.level=0)
ymat <- make_ymat(genout.infe$matches, Y)
for (gamma in seq(1, 2, 0.1)) 
  cat(sprintf("Gamma: %.1f\tp-val: %.5f\n", 
              gamma, senmv(y=ymat, gamma=gamma, method='t')$pval))
```
